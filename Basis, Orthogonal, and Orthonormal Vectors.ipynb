```python
import numpy as np
```

```python
# 1. Buat vector i dan j
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
print("Vector i:", i)
print("Vector j:", j)
```

```python
# Cek orthogonal (dot product)
dot_ij = np.dot(i, j)
print("\nDot product i · j =", dot_ij)
```

```python
if dot_ij == 0:
    print("Vector i dan j orthogonal")
else:
    print("Vector i dan j tidak orthogonal")
```

```python
# Cek apakah orthonormal (panjang = 1)
norm_i = np.linalg.norm(i)
norm_j = np.linalg.norm(j)
```

```python
print("\nPanjang i:", norm_i)
print("Panjang j:", norm_j)

if norm_i == 1 and norm_j == 1:
    print("Vector i dan j orthonormal")
```

```python
# 2. Vector v dan w
v = np.array([1, 2])
w = np.array([2, -1])
```

```python
print("\nVector v:", v)
print("Vector w:", w)
```

```python
# Cek orthogonal
dot_vw = np.dot(v, w)
print("\nDot product v · w =", dot_vw)

if dot_vw == 0:
    print("Vector v dan w orthogonal")
else:
    print("Vector v dan w tidak orthogonal")
```
