```python
import numpy as np
```

```python
# Buat matrix X ukuran 2x3
X = np.array([[1, 3, 5],
              [2, 4, 6]])
```

```python
print("Matrix X:")
print(X)
```

```python
# Tampilkan ukuran matrix
print("\nUkuran matrix X:", X.shape)
```

```python
# Kolom pertama
print("\nKolom pertama:")
print(X[:, 0])
```

```python
# Baris kedua
print("\nBaris kedua:")
print(X[1, :])
```
