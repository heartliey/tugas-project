```python
# Tampilkan tipe data dari x
y_tf2 = tf.Variable(0.82)
```

```python
# Buat variabel skalar y dan simpan suatu nilai skalar
x_tf.shape
```

```python
# Tampilkan tipe data dari y
y_tf = tf.Variable(3, dtype=tf.int16)
```

```python
# Jumlahkan x + y
x_tf + y_tf
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
# Jumlahkan x float dengan y
tf_sum.numpy() # note that NumPy operations automatically convert tensors to NumPy arrays, and vice versa
```

```python
# Tampilkan tipe data dari x float
type(tf_sum.numpy())
```

Lakukan Tahapan Pada Soal No 1 Menggunakan Pytorch

```python
import torch
```

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x = torch.tensor(17)
```

```python
# tampilkan tipe data dari x
print("Nilai x:", x)
print("Tipe data x:", x.dtype)
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y = torch.tensor(17)
```

```python
# tampilkan tipe data dari y
print("\nNilai y:", y)
print("Tipe data y:", y.dtype)
```

```python
# jumlahkan x + y
hasil = x + y
print("\nHasil x + y:", hasil)
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = torch.tensor(1.7)
```

```python
# jumlahkan x_float dengan y (perlu samakan tipe data)
y_float = y.to(torch.float32)
hasil_float = x_float + y_float
```

```python
print("\nNilai x_float:", x_float)
print("Tipe data x_float:", x_float.dtype)
```
